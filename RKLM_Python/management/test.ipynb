```python
import numpy as np
from copy import deepcopy

import logging

N = 10

# class member(object):
#     A = 10
#     B = 2
#     C = 'D'

mem = np.arange(25).reshape(5,5)
logging.info(mem)
def sampler(ic):
    # ic.A += np.random.randint(5)
    ic += np.random.randint(5)

class ensemble(object):
    def __init__(self):
        None
        
    def ensemble(self):
        for key,value in vars(self).items():
            setattr(self,key,value)

    def set_members(self,ic,N,sampler):
        cnt = 0
        for _ in range(N):
            mem = deepcopy(ic)
            sampler(mem)
            setattr(self,'mem_' + str(cnt),mem)
            cnt += 1

    def get_mean(self, ensemble):
        return self.members(ensemble).sum(axis=0)

    def get_covariance(self, ensemble):
        ensemble_mean = self.get_mean(ensemble)
        N = ensemble.members(ensemble).shape[0]
        ensemble_list = ensemble.members(ensemble)
        distance = ensemble_list - ensemble_mean
        ensemble_covar = 1./(N-1.) * (distance.transpose((0,2,1)) @ distance).sum(axis=0)
        return ensemble_covar

    @staticmethod
    def members(ensemble):
        return np.array(list(ensemble.__dict__.values()))


    # def __repr__(ensemble):
    #     return ensemble.__dict__.values()

ens = ensemble()
ens.set_members(mem,N,sampler)

```
