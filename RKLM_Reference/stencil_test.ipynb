```python
import numpy as np
```

```python
AA = np.arange(12*10).reshape(12,10)
print(AA)

iicxn = AA.shape[1]
iicyn = AA.shape[0]
print(iicxn)
print(iicyn)
```

```python
# idx = 116
# cnt_x = 5
# cnt_y = 11

cnt_x = 0
cnt_y = 0
cnt = 0

x_periodic = True

for idx in range(iicxn * iicyn):
    
    topleft = idx - iicxn - 1
    midleft = idx - 1
    botleft = idx + iicxn - 1

    topmid = idx - iicxn
    midmid = idx
    botmid = idx + iicxn

    topright = idx - iicxn + 1
    midright = idx + 1
    botright = idx + iicxn + 1

    if cnt_x == 0:
        topleft += iicxn - 1
        midleft += iicxn - 1
        botleft += iicxn - 1

        if x_periodic:
            topmid += iicxn - 1
            midmid += iicxn - 1
            botmid += iicxn - 1

    if cnt_x == iicxn - 1:
        topright -= iicxn - 1
        midright -= iicxn - 1
        botright -= iicxn - 1

        if x_periodic:
            topmid -= iicxn - 1
            midmid -= iicxn - 1
            botmid -= iicxn - 1


    if cnt_y == 0:
        topleft += (iicxn * (iicyn))
        topmid += (iicxn * (iicyn))
        topright += (iicxn * (iicyn))

    if cnt_y == iicyn - 1:
        botleft -= (iicxn * (iicyn))
        botmid -= (iicxn * (iicyn))
        botright -= (iicxn * (iicyn))
        
    stencil = [[topleft, topmid, topright],[midleft, midmid, midright],[botleft,botmid,botright]]
    stencil = np.array(stencil)
    

    print(AA.ravel()[stencil.ravel()].reshape(3,3))
    print("%d -------------------------------\n" %cnt)
    cnt += 1
    cnt_x += 1
    if cnt_x % iicxn == 0:
        cnt_y += 1
        cnt_x = 0
```

```python
def negative_symmetric(vector,pad_width,iaxis,kwargs=None):
    if pad_width[1] > 0:
        sign = -1.
        vector[:pad_width[0]] = sign * vector[pad_width[0]:2*pad_width[0]][::-1]
        vector[-pad_width[1]:] = sign * vector[-2*pad_width[1]:-pad_width[1]][::-1]
        return vector
    else: # axis must have length > 0 for padding
        return vector
```

```python
print(np.pad(AA,((0,0),(2,2)),negative_symmetric))
```
